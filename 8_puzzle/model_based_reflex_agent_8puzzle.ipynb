```python
import random

def check_solvable(arr):
    inv = 0
    for i in range(9):
        for j in range(i + 1, 9):
            if arr[i] != 0 and arr[j] != 0 and arr[i] > arr[j]:
                inv += 1
    return inv % 2 == 0

ran = random.sample(range(9), 9)
while not check_solvable(ran):
    ran = random.sample(range(9), 9)

matrix = []
i = 0
index = ()

for r in range(3):
    row = []
    for c in range(3):
        if ran[i] == 0:
            index = (r, c)
        row.append(ran[i])
        i += 1
    matrix.append(row)

print("Bàn cờ ban đầu: ")
for i in range(3):
    print(matrix[i])

print("Vị trí ban đầu: ", index, "\n")
```

```python
x = [0, -1, 0, 1]
y = [-1, 0, 1, 0]

# 0, 1, 2, 3 lần lượt là LEFT, UP, RIGHT, DOWN
def action_A(r, c):
    action = []

    if r > 0:
        action.append(1)
    if r < 2:
        action.append(3)
    if c > 0:
        action.append(0)
    if c < 2:
        action.append(2)

    return action

def get_action(his, matrix, r, c):
    action = []
    action_B = action_A(r, c)

    for i in range(len(action_B)):
        r_new, c_new = r + x[action_B[i]], c + y[action_B[i]]
        
        temp_matrix = [row for row in matrix]
        temp_matrix[r][c], temp_matrix[r_new][c_new] = temp_matrix[r_new][c_new], temp_matrix[r][c]
        state = tuple(tuple(row) for row in temp_matrix)

        if state not in his:
            action.append(action_B[i])

    if len(action) == 0:
        return action_B

    return action
```

```python
goal = [[1, 2, 3],
        [4, 5, 6],
        [7, 8, 0]]

his = set()
count = 0

while matrix != goal and count < 10000:
    r, c = index
    
    current_state = tuple(tuple(row) for row in matrix)
    his.add(current_state)

    next = random.choice(get_action(his, matrix, r, c))

    r_new, c_new = r + x[next], c + y[next]

    matrix[r][c], matrix[r_new][c_new] = matrix[r_new][c_new], matrix[r][c]
    index = (r_new, c_new)

    print("Vị trí tiếp theo là: ", index)    
    print("Ma trận sau khi hoán đổi: ")    
    for i in range(3):
        print(matrix[i])
    print("\n")

    count += 1

print("Số lượng các vị trí đã đi: ", len(his))
print(his)
```
