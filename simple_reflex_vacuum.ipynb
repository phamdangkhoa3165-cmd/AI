```python
import random

r_n = int(input("Nhập số hàng: "))
c_n = int(input("Nhập số cột: "))

b = random.choices([0,1], k = r_n * c_n)

matrix = []
trash = 0

index_vacuum = ()

i = 0
for r in range(r_n):
    row = []
    for c in range(c_n):
        if b[i] == 1:
            trash += 1

            if not index_vacuum:
                index_vacuum = (r, c)
        row.append(b[i])
        i += 1

    matrix.append(row)

for r in range(r_n):
    print(matrix[r])

print("Vị trí máy hút bụi hiện tại: ", index_vacuum)
```

```python
x = [0, -1, 0, 1]
y = [-1, 0, 1, 0]
move = ["LEFT", "UP", "RIGHT", "DOWN"]

action = []

# 0, 1, 2, 3 lần lượt là LEFT, UP, RIGHT, DOWN
def get_action(x, y):
    action = []

    if x > 0:
        action.append(1)
    if x < r_n - 1:
        action.append(3)
    if y > 0:
        action.append(0)
    if y < c_n - 1:
        action.append(2)

    return action
```

```python
while trash > 0:
    action = 0

    r, c = index_vacuum

    if matrix[r][c] == 1:
        print("Đã hút xong bụi!")
        matrix[r][c] = 0
        action = random.choice(get_action(r, c))

        trash -= 1

    elif matrix[r][c] == 0:
        print("Ở đây sạch!")
        action = random.choice(get_action(r, c))

    print("Hành động tiếp theo là: ", move[action])

    index_vacuum = (r + x[action], c + y[action])

    print("Vị trí tiếp theo là: ", index_vacuum)

    print("Sàn nhà sau khi máy hút bụi di chuyển:")
    for i in range(r_n):
        print(matrix[i])

    print("\n")
```
