```python
import random

r_n = int(input("Nhập số hàng: "))
c_n = int(input("Nhập số cột: "))

b = random.choices([0,1], k = r_n * c_n)

matrix = []
goal = []

i = 0
for r in range(r_n):
    row = []
    goal_r = []
    for c in range(c_n):
        row.append(b[i])
        i += 1

        goal_r.append(0)

    matrix.append(row)

    goal.append(goal_r)

for r in range(r_n):
    print(matrix[r])
```

```python
vacuum_index = (0, 0)

x = [0, -1, 0, 1]
y = [-1, 0, 1, 0]

his = [(0,0)] 

while matrix != goal:
    r, c = vacuum_index

    if (r,c) not in his:
        his.append((r,c))

    if matrix[r][c] == 1:
        matrix[r][c] = 0

    valid = []
    fallback_valid = []
    
    for i in range(4):
        if 0 <= r + x[i] < r_n and 0 <= c + y[i] < c_n:
            fallback_valid.append(i)

            if (r + x[i], c + y[i]) not in his:
                valid.append(i)

    print(valid)

    if len(valid) == 0:
        next_step = random.choice(fallback_valid)
    else:
        next_step = random.choice(valid)

    vacuum_index = (r + x[next_step], c + y[next_step])

    print(vacuum_index)

    print(his)

    for i in range(r_n):
        print(matrix[i])
```
