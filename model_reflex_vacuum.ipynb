```python
import random

r_n = int(input("Nhập số hàng: "))
c_n = int(input("Nhập số cột: "))

b = random.choices([0,1], k = r_n * c_n)

matrix = []
one_count = 0

i = 0
for r in range(r_n):
    row = []
    for c in range(c_n):
        if b[i] == 1:
            one_count += 1
        row.append(b[i])
        i += 1

    matrix.append(row)

for r in range(r_n):
    print(matrix[r])

print(one_count)
```

```python
x = [0, -1, 0, 1]
y = [-1, 0, 1, 0]

def action_A(r, c):
    action = []

    if r > 0:
        action.append(1)
    if r < r_n - 1:
        action.append(3)
    if c > 0:
        action.append(0)
    if c < c_n - 1:
        action.append(2)

    return action

def get_action(his, r, c):
    action = []

    action_B = action_A(r,c)

    for i in range(len(action_B)):
        r_new, c_new = r + x[action_B[i]], c + y[action_B[i]]
        if (r_new, c_new) not in his:
            his.add((r_new, c_new))

            action.append(action_B[i])

    if len(action) == 0:
        return action_B
    
    return action
```

```python
vacuum_index = (0, 0)

his = set()

step_count = 0

while one_count > 0:
    r, c = vacuum_index

    his.add((r,c))

    if matrix[r][c] == 1:
        matrix[r][c] = 0

        one_count -= 1

        next = random.choice(get_action(his, r, c))

        vacuum_index = (r + x[next], c + y[next])
    else:
        next = random.choice(get_action(his, r, c))

        vacuum_index = (r + x[next], c + y[next])

    print(vacuum_index)

    for i in range(r_n):
        print(matrix[i])

    step_count += 1
```

```python
print(step_count)
print(len(his))
print(his)
```
