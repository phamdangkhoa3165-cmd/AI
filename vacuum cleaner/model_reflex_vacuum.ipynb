```python
import random

r_n = int(input("Nhập số hàng: "))
c_n = int(input("Nhập số cột: "))

b = random.choices([0,1], k = r_n * c_n)

matrix = []
one_count = 0

i = 0
for r in range(r_n):
    row = []
    for c in range(c_n):
        if b[i] == 1:
            one_count += 1
        row.append(b[i])
        i += 1

    matrix.append(row)

for r in range(r_n):
    print(matrix[r])
```

```python
x = [0, -1, 0, 1]
y = [-1, 0, 1, 0]
```

```python
vacuum_index = (0, 0)

his = set()

step_count = 0

while one_count > 0:
    r, c = vacuum_index

    his.add((r,c))

    if matrix[r][c] == 1:
        matrix[r][c] = 0

        one_count -= 1

    valid = []
    fallback_valid = []
    
    for i in range(4):
        if 0 <= r + x[i] < r_n and 0 <= c + y[i] < c_n:
            fallback_valid.append(i)

            if (r + x[i], c + y[i]) not in his:
                valid.append(i)

    if len(valid) == 0:
        next_step = random.choice(fallback_valid)
    else:
        next_step = random.choice(valid)

    vacuum_index = (r + x[next_step], c + y[next_step])

    print(vacuum_index)

    for i in range(r_n):
        print(matrix[i])

    step_count += 1

print(step_count)
```
