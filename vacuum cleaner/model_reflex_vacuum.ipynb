```python
import random

r_n = int(input("Nhập số hàng: "))
c_n = int(input("Nhập số cột: "))

b = random.choices([0,1], k = r_n * c_n)

matrix = []
one_count = 0

index = ()

i = 0
for r in range(r_n):
    row = []
    for c in range(c_n):
        if b[i] == 1:
            one_count += 1

            if not index:
                index = (r, c)

        row.append(b[i])
        i += 1

    matrix.append(row)

for r in range(r_n):
    print(matrix[r])

print("Vị trí hiện tại của máy hút bụi là: ", index)
```

```python
x = [0, -1, 0, 1]
y = [-1, 0, 1, 0]

# 0, 1, 2, 3 lần lượt là: LEFT, UP, RIGHT, DOWN

def get_action_A(x, y):
    action = []

    if x > 0:
        action.append(1)
    if x < r_n - 1:
        action.append(3)
    if y > 0:
        action.append(0)
    if y < c_n - 1:
        action.append(2)

    return action

def get_action(his, r, c):
    action = []

    action_B = get_action_A(r, c)

    for i in range(len(action_B)):
        r_new, c_new = r + x[action_B[i]], c + y[action_B[i]]

        if (r_new, c_new) not in his:
            his.add((r_new, c_new))

            action.append(action_B[i])

    if len(action) == 0:
        return action_B
    
    return action
```

```python
his = set()

his.add(index)

step_count = 0

while one_count > 0:
    r, c = index

    if matrix[r][c] == 1:
        print("Vị trí này bẩn! Hút bụi!")
        matrix[r][c] = 0

        next = random.choice(get_action(his, r, c))

        index = (r + x[next], c + y[next])

        one_count -= 1
    else:
        print("Vị trí này sạch!")
        next = random.choice(get_action(his, r, c))

        index = (r + x[next], c + y[next])

    step_count += 1

    print("Sàn nhà sau khi máy hút bụi đi qua:")

    for i in range(r_n):
        print(matrix[i])

    print("Máy hút bụi sẽ đi đến ô: ", index)

    print("\n")

print("Số bước di chuyển của máy hút bụi: ", step_count)

print("Chiều dài số bước đã đi: ", len(his))

print(his)
```
